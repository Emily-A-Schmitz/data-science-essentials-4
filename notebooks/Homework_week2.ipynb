```python
import pandas as pd
physicians = pd.read_csv('../data/primary_care_physicians.csv')
physicians
```

```python
physicians = physicians.loc[physicians["state"] == "Tennessee"]
physicians
```

```python
physicians.shape
```

```python
physicians['state'] = physicians['state'].str.replace('Tennessee', 'tennessee')
physicians
```

```python
population = pd.read_csv('../data/population_by_county.csv')
population
```

```python
population = population.loc[population["state"] == "TENNESSEE"]
population.shape
```

```python
population['state'] = population['state'].str.lower()
population['county'] = population['county'].str.split(' ', expand = True)[0]
population
```

```python
physicians = pd.merge(physicians, population)
physicians
```

```python
physicians['urban'].value_counts()
```

```python
adequate_counties = physicians.loc[(physicians['population'] / physicians['primary_care_physicians']) < 1500]
adequate_counties.shape
```

```python
moderately_inadequate = physicians.loc[((physicians['population'] / physicians['primary_care_physicians']) >= 1500) & ((physicians['population'] / physicians['primary_care_physicians'])<3500)]
moderately_inadequate.shape
```

```python
low_inadequate = physicians.loc[(physicians['population'] / physicians['primary_care_physicians']) >= 3500]
low_inadequate.shape
```
